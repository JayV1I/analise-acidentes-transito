### 1. Upload das Planilhas
Por favor, faça o upload da sua planilha "antiga" e da sua planilha "nova".

```python
from google.colab import files
import pandas as pd

print("Faça o upload da planilha antiga (ex: antiga.csv):")
uploaded_old = files.upload()
old_file_name = next(iter(uploaded_old))

print("\nFaça o upload da planilha nova (ex: nova.csv):")
uploaded_new = files.upload()
new_file_name = next(iter(uploaded_new))

print(f"\nPlanilha antiga: {old_file_name}")
print(f"Planilha nova: {new_file_name}")
```

### 2. Leitura e Unificação das Planilhas
Agora, vamos ler os arquivos e uní-los. Como os arquivos são Excel, usaremos `pd.read_excel`.

```python
# Lendo a planilha antiga (agora usando read_excel para arquivos .xlsx)
df_old = pd.read_excel(old_file_name)

# Lendo a planilha nova (agora usando read_excel para arquivos .xlsx)
df_new = pd.read_excel(new_file_name)

# Unindo as planilhas sem sobreposição (concatenando verticalmente)
df_unified = pd.concat([df_old, df_new], ignore_index=True)

print("\nPrimeiras 5 linhas da planilha unificada:")
display(df_unified.head())
```

### 3. Salvar e Baixar a Planilha Unificada
Por fim, você pode baixar a nova planilha unificada.

```python
output_file_name = 'planilha_unificada.xlsx'
df_unified.to_excel(output_file_name, index=False)

print(f"\nPlanilha unificada salva como '{output_file_name}'.")
files.download(output_file_name)
```

### 4. Criando um Aplicativo Tkinter para Unificar Planilhas

Para criar um aplicativo de desktop com uma interface gráfica (GUI) usando Tkinter, você precisará de um código Python que possa ser executado em seu ambiente local. O código a seguir pode ser copiado e salvo como um arquivo `.py` (por exemplo, `unificador_planilhas.py`) e executado a partir do seu terminal.

**Como executar:**
1.  Copie todo o código da próxima célula.
2.  Cole-o em um editor de texto e salve o arquivo como `unificador_planilhas.py`.
3.  Abra um terminal ou prompt de comando na mesma pasta onde você salvou o arquivo.
4.  Execute o comando: `python unificador_planilhas.py`

Isso abrirá uma janela do aplicativo onde você poderá selecionar suas planilhas e unificá-las.

import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd

class UnificadorPlanilhasApp:
    def __init__(self, master):
        self.master = master
        master.title("Unificador de Planilhas")

        self.label = tk.Label(master, text="Unificador de Planilhas XLSX")
        self.label.pack(pady=10)

        self.old_file_path = ""
        self.new_file_path = ""

        self.btn_select_old = tk.Button(master, text="Selecionar Planilha Antiga", command=self.select_old_file)
        self.btn_select_old.pack(pady=5)
        self.lbl_old_file = tk.Label(master, text="Nenhuma planilha antiga selecionada.")
        self.lbl_old_file.pack()

        self.btn_select_new = tk.Button(master, text="Selecionar Planilha Nova", command=self.select_new_file)
        self.btn_select_new.pack(pady=5)
        self.lbl_new_file = tk.Label(master, text="Nenhuma planilha nova selecionada.")
        self.lbl_new_file.pack()

        self.btn_unify = tk.Button(master, text="Unificar Planilhas", command=self.unify_spreadsheets)
        self.btn_unify.pack(pady=20)

        self.status_label = tk.Label(master, text="")
        self.status_label.pack(pady=10)

    def select_old_file(self):
        self.old_file_path = filedialog.askopenfilename(
            title="Selecione a Planilha Antiga",
            filetypes=[("Arquivos Excel", "*.xlsx *.xls")]
        )
        if self.old_file_path:
            self.lbl_old_file.config(text=f"Antiga: {self.old_file_path.split('/')[-1]}")
        else:
            self.lbl_old_file.config(text="Nenhuma planilha antiga selecionada.")

    def select_new_file(self):
        self.new_file_path = filedialog.askopenfilename(
            title="Selecione a Planilha Nova",
            filetypes=[("Arquivos Excel", "*.xlsx *.xls")]
        )
        if self.new_file_path:
            self.lbl_new_file.config(text=f"Nova: {self.new_file_path.split('/')[-1]}")
        else:
            self.lbl_new_file.config(text="Nenhuma planilha nova selecionada.")

    def unify_spreadsheets(self):
        if not self.old_file_path or not self.new_file_path:
            messagebox.showwarning("Atenção", "Por favor, selecione ambas as planilhas.")
            return

        self.status_label.config(text="Unificando planilhas...")
        self.master.update_idletasks() # Força a atualização da GUI

        try:
            df_old = pd.read_excel(self.old_file_path)
            df_new = pd.read_excel(self.new_file_path)

            df_unified = pd.concat([df_old, df_new], ignore_index=True)

            output_file_name = filedialog.asksaveasfilename(
                defaultextension=".xlsx",
                filetypes=[("Arquivos Excel", "*.xlsx")],
                title="Salvar Planilha Unificada Como"
            )

            if output_file_name:
                df_unified.to_excel(output_file_name, index=False)
                self.status_label.config(text=f"Planilha unificada salva em: {output_file_name.split('/')[-1]}")
                messagebox.showinfo("Sucesso", "Planilhas unificadas e salvas com sucesso!")
            else:
                self.status_label.config(text="Operação cancelada.")

        except Exception as e:
            self.status_label.config(text=f"Erro: {e}")
            messagebox.showerror("Erro", f"Ocorreu um erro ao unificar as planilhas: {e}")

if __name__ == "__main__":
    root = tk.Tk()
    app = UnificadorPlanilhasApp(root)
    root.mainloop()
